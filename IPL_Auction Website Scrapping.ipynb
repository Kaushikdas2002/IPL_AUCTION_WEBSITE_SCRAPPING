# IPL AUCTION WEBSITE SCRAPPING

### Importing the required libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

### Reading the website

```python
url = 'https://www.iplt20.com/auction/2024'
r = requests.get(url)

print(r.status_code)
```

```python
soup = BeautifulSoup(r.text,'lxml')
```

### Reading the table from the HTML code

```python
table = soup.find('table',class_ = 'ih-td-tab auction-tbl')
print(table)
```

### Finding the Headers

```python
header = table.find_all("th")

title = []

for i in header:
    name = i.text
    title.append(name)

print(title)
```

```python
df = pd.DataFrame(columns = title)
df
```

### Adding the data to the dataframe

```python
rows = table.find_all('tr')

for i in rows[1:]:
    print(i.text)
```

```python
for i in rows[1:]:
    data = i.find_all('td')
    row = [tr.text.strip() for tr in data]
    print(row)
    
    l = len(df)
    df.loc[l] = row
```
